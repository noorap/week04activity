```python
# import libraries and load the data to python environment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel('Telco-Customer-Churn.xlsx')
df
```

# 1. Compare churn count with respect to gender

```python
sns.countplot(x='gender',data=df,hue='Churn')
plt.title('Gender Vs Churn count')
```

# 2.Find out how many female senior citizens there in the dataset

```python
# To find the count of female senior citizens
df[(df['gender']=='Female') &  (df['SeniorCitizen']==1.0)]['SeniorCitizen'].count()
```

```python
# Visualization of the count of female senior citizens
df2=df[(df['gender']=='Female') & (df['SeniorCitizen']==1.0)]
sns.countplot(x='SeniorCitizen',data=df2, color='green')
```

#  3.Compare 'tenure' with 'Total Charges'

```python
sns.scatterplot(data=df,x='tenure',y='TotalCharges')
plt.title('Tenure Vs Total Charges')
plt.tight_layout()
```

# 4.Find out which contract is preferred by the senior citizen.

```python
df1 = df[df['SeniorCitizen']==1.0]
sns.countplot(x ='Contract', data = df1)
plt.title("Contracts preferred by Senior Citizens")
plt.tight_layout()
```

```python
# Histogram of Payment Methods
plt.figure(figsize=(8,6))
plt.hist(df['PaymentMethod'],color='green')
plt.title('Histogram of Payment Method')
plt.xlabel('PaymentMethod')
plt.ylabel('Frequency')
```
